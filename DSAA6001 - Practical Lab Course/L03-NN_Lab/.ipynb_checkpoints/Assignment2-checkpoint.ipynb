````python
# Define a global transformer to appropriately scale images and subsequently convert them to a Tensor.
img_size = 224
loader = transforms.Compose([
  transforms.Resize(img_size),
  transforms.CenterCrop(img_size),
  transforms.ToTensor(),
])

def load_image(filename):
    """
    Simple function to load and preprocess the image.

    1. Open the image.
    2. Scale/crop it and convert it to a float tensor.
    3. Convert it to a variable (all inputs to PyTorch models must be variables).
    4. Add another dimension to the start of the Tensor (b/c VGG expects a batch).
    5. Move the variable onto the GPU.
    """
    image = PILImage.open(filename).convert('RGB')
    image_tensor = loader(image).float()
    image_var = Variable(image_tensor).unsqueeze(0)
    return image_var


load_image('data/data_sampled/train2014/COCO_train2014_000000000009.jpg')
````

````python
# Load ImageNet label to category name mapping.
imagenet_categories = [value for key,value in sorted(json.load(open('data/data_sampled/imagenet_categories.json')).items(), key=lambda t: int(t[0]))]

# Load annotations file for the 20K training images.
mscoco_train = json.load(open('data/data_sampled/annotations/train2014.json'))
train_ids = [entry['id'] for entry in mscoco_train['images']]
train_id_to_file = {entry['id']: 'data/data_sampled/train2014/' + entry['file_name'] for entry in mscoco_train['images']}
category_to_name = {entry['id']: entry['name'] for entry in mscoco_train['categories']}
category_idx_to_name = [entry['name'] for entry in mscoco_train['categories']]
category_to_idx = {entry['id']: i for i,entry in enumerate(mscoco_train['categories'])}

# Load annotations file for the 100 validation images.
mscoco_val = json.load(open('data/data_sampled/annotations/val2014.json'))
val_ids = [entry['id'] for entry in mscoco_val['images']]
val_id_to_file = {entry['id']: 'data/data_sampled/val2014/' + entry['file_name'] for entry in mscoco_val['images']}

# We extract out all of the category labels for the images in the training set. We use a set to ignore
# duplicate labels.
train_id_to_categories = defaultdict(set)
for entry in mscoco_train['annotations']:
    train_id_to_categories[entry['image_id']].add(entry['category_id'])

# We extract out all of the category labels for the images in the validation set. We use a set to ignore
# duplicate labels.
val_id_to_categories = defaultdict(set)
for entry in mscoco_val['annotations']:
    val_id_to_categories[entry['image_id']].add(entry['category_id'])

from IPython.display import Image, display
display(Image(filename='data/data_sampled/val2014/COCO_val2014_000000391895.jpg'))
````

# Making Predictions Using VGG-16

Given the pre-trained network, we must now write the code to make predictions on the 10 validation images via a forward pass through the network. Typically the final layer of VGG-16 is a softmax layer, however the pre-trained PyTorch model that we are using does not have softmax built into the final layer (instead opting to incorporate it into the loss function) and therefore we must **manually** apply softmax to the output of the function.

````python
softmax = nn.Softmax()
for image_id in val_ids[:10]:
    # Display the image.
    # -- Your code goes here --
    image_filename = val_id_to_file[image_id]
    display(Image(filename=image_filename))

    # Print all of the category labels for this image.
    # -- Your code goes here --
    print("Category labels for this image:")
    for category_id in val_id_to_categories[image_id]:
        print(f"{category_to_name[category_id]} ({category_id})")


    # Load/preprocess the image.
    img = load_image(val_id_to_file[image_id])

    # Run the image through the model and softmax.
    img = img.cuda()
    label_likelihoods = softmax(vgg_model(img)).squeeze()

    # Get the top 5 labels, and their corresponding likelihoods.
    probs, indices = label_likelihoods.topk(5)

    # Iterate and print out the predictions.
    # -- Your code goes here --
    print("\nTop 5 predicted labels:")
    for i in range(5):
        label_idx = indices[i].item()
        label_name = imagenet_categories[label_idx]
        likelihood = probs[i].item()
        print(f"{label_name} (Index {label_idx}): Likelihood = {likelihood:.4f}")

    print("\n" + "=" * 50 + "\n")  # Separator between images
````

# Computing Generic Visual Features using CNN

Since, rather than the output of VGG, we want a fixed sized vector representation of each image, we remove the last linear layer. The implementation of the forward function for VGG is shown below:

```
x = self.features(x)
x = x.view(x.size(0), -1)
x = self.classifier(x)
```
We aim to preserve everything but the final component of the classifier, meaning we must define an alternative equivalent to `self.classifier`.

````python
# Remove the final layer of the classifier, and indicate to PyTorch that the model is being used for inference
# rather than training (most importantly, this disables dropout).

# -- Your code goes here --
print(vgg_model.classifier)
new_classifier = nn.Sequential(*list(vgg_model.classifier.children())[:-1])
print(new_classifier)
# Set the model to evaluation mode to disable dropout
vgg_model.eval()
# Replace the classifier in the original VGG model
vgg_model.classifier = new_classifier
print(vgg_model)
````

````python
# First we vectorize all of the features of training images and write the results to a file.

# -- Your code goes here --
training_vectors = []

for image_id in train_ids:
    img_filename = train_id_to_file[image_id]
    img_tensor = load_image(img_filename)

    with torch.no_grad():
        features = vgg_model(img_tensor).squeeze().cpu().numpy()

    training_vectors.append(features)

# Save training feature vectors to a file
# Create the 'outputs' directory if it doesn't exist
output_dir = '/content/data/output'
os.makedirs(output_dir, exist_ok=True)
np.save(open('/content/data/output/training_vectors.npy', 'wb+'), training_vectors)
#np.save(open('/content/drive/MyDrive/w4-nn&cv/output/training_vectors.npy', 'wb+'), validation_vectors)
````

````python
# Next we vectorize all of the features of validation images and write the results to a file.

# -- Your code goes here --
validation_vectors = []
for image_id in val_ids:
    img_filename = val_id_to_file[image_id]
    img_tensor = load_image(img_filename)

    with torch.no_grad():
        features = vgg_model(img_tensor).squeeze().cpu().numpy()

    validation_vectors.append(features)
output_dir = '/content/data/output'
os.makedirs(output_dir, exist_ok=True)
np.save(open('/content/data/validation_vectors.npy', 'wb+'), validation_vectors)
#np.save(open('/content/drive/MyDrive/w4-nn&cv/output/validation_vectors.npy', 'wb+'), validation_vectors)
````

# Training a Multi-Label Classification Network

We now build a two layer classification network, which takes 4096-dimensional vectors as input and outputs the probabilities of the 80 categories present in MSCOCO.

For this purpose, we utilize two layers (both containing sigmoid activation functions) with the hidden dimension set to 512.

````python
# Use 'train_id_to_categories' to generate the 'training_labels' && Use 'val_id_to_categories' to generate the 'val_labels'
# Define the number of categories in the MSCOCO dataset
num_categories = len(category_idx_to_name)  # Since there are 80 categories in COCO dataset

# Show the mapping from original category IDs to continuous IDs
print(category_to_idx)
# Initialize an empty list to store the training and val labels
training_labels = []
val_labels = []

# Iterate through each training image
for image_id in train_ids:
    # Initialize a binary label vector with zeros for all categories
    label_vector = [0] * num_categories

    # Get the set of category IDs for this training image
    categories = train_id_to_categories[image_id]

    # Map the original category IDs to continuous IDs and set the corresponding elements in the label vector to 1
    for original_category_id in categories:
        continuous_category_id = category_to_idx[original_category_id]
        label_vector[continuous_category_id] = 1

    # Append the binary label vector to the training_labels list
    training_labels.append(label_vector)

# Iterate through each val image
for image_id in val_ids:
    # Initialize a binary label vector with zeros for all categories
    label_vector = [0] * num_categories

    # Get the set of category IDs for this val image
    categories = val_id_to_categories[image_id]

    # Map the original category IDs to continuous IDs and set the corresponding elements in the label vector to 1
    for original_category_id in categories:
        continuous_category_id = category_to_idx[original_category_id]
        label_vector[continuous_category_id] = 1

    # Append the binary label vector to the val_labels list
    val_labels.append(label_vector)
````

````python
# First we construct a class for the model
# -- Your code goes here --
# Define the classification model
class MultiLabelClassificationModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(MultiLabelClassificationModel, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.sigmoid = nn.Sigmoid()
        self.fc2 = nn.Linear(hidden_dim, output_dim)
    def forward(self, x):
        x = self.fc1(x)
        x = self.sigmoid(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x

# Initialize the multi-label classification model
input_dim = 4096  # Dimension of the feature vectors
hidden_dim = 512  # Hidden layer dimension
output_dim = 80  # Number of output categories in MSCOCO

model = MultiLabelClassificationModel(input_dim, hidden_dim, output_dim)
````

````python
# The output data is prepared by representing each output as a binary vector of categories
# -- Your code goes here --
# Load the training feature vectors
training_vectors = np.load('/content/drive/MyDrive/w4-nn&cv/output/training_vectors.npy')
validation_vectors = np.load('/content/drive/MyDrive/w4-nn&cv/output/validation_vectors.npy')


# Define the training function
def train(model, learning_rate=0.001, batch_size=100, epochs=20):
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

    # Assuming you have training_vectors and training_labels loaded and preprocessed
    # training_vectors is a list of feature vectors
    for epoch in range(epochs):
        for i in range(0, len(training_vectors), batch_size):
            inputs = torch.tensor(training_vectors[i:i+batch_size], dtype=torch.float32)
            labels = torch.tensor(training_labels[i:i+batch_size], dtype=torch.float32)
            optimizer.zero_grad()

            outputs = model(inputs)
            loss = criterion(outputs, labels)
            # print(outputs)
            # print(labels)
            loss.backward()
            optimizer.step()

        print(f"Epoch [{epoch+1}/{epochs}] Loss: {loss.item()}")
# Finally train the model
train(model)
````

````python
# Test the accuracy
# -- Your code goes here --
# Set the model to evaluation mode
model.eval()
# Convert labels to tensor
val_labels_tensor = torch.tensor(val_labels)

# Convert predictions to binary labels (using a threshold)
with torch.no_grad():
    val_predictions = model(torch.tensor(validation_vectors, dtype=torch.float32))
threshold = 0.021
val_predicted_labels = (val_predictions > threshold).float()

# Calculate accuracy
correct = (val_predicted_labels * val_labels_tensor).sum(1) / (val_labels_tensor.sum(1) + 1e-10) # Num(Both Pre-class and Label-class are 1) / Num(1-Label-class)
accuracy = correct.mean().item()

print(f"Accuracy: {correct}%")
print(f"Validation Accuracy: {accuracy*100:.2f}%")
````

# End-to-End Model Fine-tuning

Instead of training *only* the final two layers, we now create an end-to-end model and train the entire thing.

````python
# First we construct a class for the model
# -- Your code goes here --

# Define the classification model
# Vgg-feture-extractor
new_classifier = nn.Sequential(*list(vgg_model.classifier.children())[:-1])
vgg_model.eval()
vgg_model.classifier = new_classifier
# 2-layer-classifier
class MultiLabelClassificationModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(MultiLabelClassificationModel, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.sigmoid = nn.Sigmoid()
        self.fc2 = nn.Linear(hidden_dim, output_dim)
    def forward(self, x):
        x = self.fc1(x)
        x = self.sigmoid(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x
input_dim = 4096  # Dimension of the feature vectors
hidden_dim = 512  # Hidden layer dimension
output_dim = 80  # Number of output categories in MSCOCO
classification_model = MultiLabelClassificationModel(input_dim, hidden_dim, output_dim)

class CombinedModel(nn.Module):
    def __init__(self, vgg_model, classification_model):
        super(CombinedModel, self).__init__()
        self.vgg_model = vgg_model
        self.classification_model = classification_model

    def forward(self, x):
        features = self.vgg_model(x)
        output = self.classification_model(features)
        return output

# Create the combined model
combined_model = CombinedModel(vgg_model, classification_model).to(device)
````

````python
1# Define the batch size and the number of epochs for training
def train_end2end(combined_model, training_tensors, training_labels, batch_size=32, num_epochs=10, learning_rate=0.001, device="cuda"):
    # Convert the training data to PyTorch tensors
    training_tensors = torch.stack(training_tensors)
    training_labels = torch.tensor(training_labels, dtype=torch.float32)

    # Create a DataLoader for training data
    train_dataset = torch.utils.data.TensorDataset(training_tensors, training_labels)
    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

    # Define the loss function (Binary Cross-Entropy) and optimizer (e.g., Adam)
    criterion = nn.BCELoss()  # Binary Cross-Entropy Loss
    optimizer = torch.optim.Adam(combined_model.parameters(), lr=learning_rate)

    # Move the model to the specified device (e.g., "cuda" for GPU)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    combined_model.to(device)

    # Training loop
    for epoch in range(num_epochs):
        running_loss = 0.0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device).squeeze(), labels.to(device)

            # Zero the parameter gradients
            optimizer.zero_grad()

            # Forward pass
            outputs = combined_model(inputs)

            # Compute the loss
            loss = criterion(outputs, labels)

            # Backpropagation and optimization
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        # Print the average loss for this epoch
        print(f"Epoch {epoch+1}/{num_epochs}, Loss: {running_loss/len(train_loader)}")

    print("Training finished!")
    return combined_model

train_end2end(combined_model, training_tensors, training_labels, batch_size=32, num_epochs=20, learning_rate=0.001, device="cuda")
````

# Hyper-parameter Tuning

Now we do a grid search over the learning rate and batch size.

````python
# -- Your code goes here --
import gc
gc.collect() # Python thing
# torch.cuda.empty_cache() # PyTorch thing
with torch.no_grad():
    torch.cuda.empty_cache()

import itertools

# Define a range of learning rates and batch sizes to search
learning_rates = [0.001, 0.01, 0.1]
batch_sizes = [32, 64, 128]

best_accuracy = 0.0  # Initialize the best accuracy to zero
best_lr = None       # Initialize the best learning rate to None
best_bs = None       # Initialize the best batch size to None

# Iterate over all combinations of learning rates and batch sizes
for lr, bs in itertools.product(learning_rates, batch_sizes):
    # Create a new instance of the model for each combination
    model = combined_model  # Replace with your model initialization code

    # Train the model with the current combination of lr and bs
    ########### !!!!!!!!To save time, just use the 200 of trainset to display
    train_end2end(model, training_tensors[:200], training_labels[:200], batch_size=bs, num_epochs=20, learning_rate=lr, device="cuda")

    # If the current combination has a higher accuracy, update the best values
    # Convert predictions to binary labels (using a threshold)
    with torch.no_grad():
      val_predictions = combined_model(torch.stack(val_tensors).squeeze())
    threshold = 0.021
    val_predicted_labels = (val_predictions > threshold).float()

    # Calculate accuracy
    correct = (val_predicted_labels * val_labels_tensor).sum(1) / (val_labels_tensor.sum(1) + 1e-10) # Num(Both Pre-class and Label-class are 1) / Num(1-Label-class)
    accuracy = correct.mean().item()
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_lr = lr
        best_bs = bs

# Print the best hyperparameters and accuracy
print(f"Best Learning Rate: {best_lr}")
print(f"Best Batch Size: {best_bs}")
print(f"Best Validation Accuracy: {best_accuracy*100:.2f}%")
````
